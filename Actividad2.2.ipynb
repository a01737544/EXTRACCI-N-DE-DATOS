```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Convertir en dataframe
df= pd.read_csv('Actividad 1 Entendimiento de los Datos.csv')
df
```

Los registros de 5 “host name” diferentes


```python
#  a) Primeros 5 host_name diferentes
cinco_hosts = df["property_type"].unique()
cinco_hosts
```

```python
tipos_elegidos = ["Room in boutique hotel", "Entire townhouse", "Entire rental unit", "Entire guest suite", "Private room in home",] 
a_result = df[df["property_type"].isin(tipos_elegidos)]
a_result
df1 = a_result.groupby(['property_type'])['property_type'].count()
df1
```

```python
f1=a_result.to_csv("fltro01.csv", index=False)
```

Los registros de los host que se hallan unido a Airbnb “despues del año 2020”

```python
# b) Host_since > 2020
df["host_since"] = pd.to_datetime(df["host_since"], errors="coerce")
b_result = df[df["host_since"] >= "2021-01-01"]
b_result
```

```python
df.info()
```

```python
df2 = b_result.groupby(['host_since'])['host_since'].count()
df2
```

```python
f2=b_result.to_csv("filtro02.csv", index=False)   
```

Los registros de los host que responden “a más tardar en 1 día"

```python
# c) Respuesta rápida (en 1 día o menos)
df3 = df.groupby(['host_response_time'])['host_response_time'].count()
df3
```

Los registros de las filas “Múltiplos de 200… (400, 600, 800, etc.

```python
# c) Respuesta rápida (en 1 día o menos)
rapidas = ["within an hour", "within a few hours", "within a day"]
c_result = df[df["host_response_time"].str.lower().isin(rapidas)]
c_result
```

```python
f3=c_result.to_csv("filtro03.csv", index=False)
```

Los registros de las filas “Múltiplos de 200… (400, 600, 800, etc.)”

```python
filas= df.iloc[0::200,:] 
filas
```

```python
f4=filas.to_csv("filtro04.csv", index=False)
```

Los registros de los tipos de cuarto: “Entire home” y “Private room”

```python
df4 = df.groupby(['room_type'])['room_type'].count()
df4
```

```python
# e) Filtrar tipos de cuarto específicos
permitidos = ["entire home/apt", "private room"]
e_result = df[df["room_type"].str.lower().isin(permitidos)]
e_result
```

```python
f5=e_result.to_csv("filtro05.csv", index=False)
```

Los registros que “a lo mucho cuenten con 4 camas”

```python
df5 = df.groupby(['beds'])['beds'].count()
df5
```

```python
# f) Filtro por número de camas
df["beds"] = pd.to_numeric(df["beds"], errors="coerce")
f_result = df[df["beds"] <= 4]
f_result
```

```python
f6=f_result.to_csv("filtro05.csv", index=False)
```

Los registros que “sean superhost”

```python
df6 = df.groupby(['host_is_superhost'])['host_is_superhost'].count()
df6
```

```python
# g) Filtro por superhost
g_result = df[df["host_is_superhost"] == "t"]   # o True si ya está como boolean
g_result
```

```python
f7=g_result.to_csv("filtro07.csv", index=False)
```

Los registros de las columnas “impares”

```python
# h) Columnas en posiciones impares (1, 3, 5…)
odd_cols = [col for i, col in enumerate(df.columns, start=1) if i % 2 == 1]
h_result = df[odd_cols]
h_result
```

```python
f8=h_result.to_csv("filtro08.csv", index=False)
```
