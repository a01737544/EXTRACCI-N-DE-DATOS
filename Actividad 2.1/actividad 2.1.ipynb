```python
#Importamos librer√≠as requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Convertir en dataframe
df= pd.read_csv('Datos_Facturaci√≥n.csv')
df
```

```python
#CVE_CLPV      (valores: 1000 a 2000)
Filtro1 = df[(df["CVE_CLPV"] >= 1000) & (df["CVE_CLPV"] <= 2000)]
Filtro1
```

```python
#CVE_VEND     (Todas las claves excepto ‚Äú5‚Äù y ‚Äú4‚Äù)
Filtro2 = df[~df["CVE_VEND"].isin([4,5])]
Filtro2
```

```python
#FECHA_ENT   (Fechas de ‚Äú28/02/2022‚Äù)
Filtro3 = df[df["FECHA_ENT"] == '2022-02-28 00:00:00']
Filtro3
```

```python
#CAN_TOT       (Cantidades menores a 5951.7)   o  STATUS (‚ÄúE‚Äù)
Filtro4= df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "E")]
Filtro4
```

```python
#Solo las columnas: CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT
Filtro5 = df[["CVE_DOC","FECHA_ENT","FECHA_VEN","CAN_TOT"]]
Filtro5
```

```python
#Solo las filas de 7001-7099
Filtro6 = df.iloc[7000:7100,:]   
Filtro6
```

```python
df1= pd.read_csv('Datos_Facturaci√≥n.csv', index_col=4)
df1
```

```python
#Index= CVE_VEND (valores: 1, 2)  (columna: FECHAELAB)
filtro7=df1.loc[[1,2], ['FECHAELAB']]
filtro7
```
